**Part 1: output with console.log()**

console.log is built into TypeScript to print out the values of variables, for example

```typescript
const title = 'TypeScript Tutorial';
const count = 25;
const isEnabled = true;

console.log(title);
console.log(count);
console.log(isEnabled);
```

**PART 3: String quotes**

Strings can be created using single quotes `'`, double quotes `"`, or backticks `` ` ``. 

All three create the same `string` type. 



```typescript
const singleQuote = 'Hello World';
const doubleQuote = "Hello World";
const backtickQuote = `Hello World`;

console.log(singleQuote);
console.log(doubleQuote);
console.log(backtickQuote);
```

**Part 4: Embedding Expressions in Strings**

In JavaScript, we call anything that produces a value an **expression**

Strings which are surrounded with backticks have an additional feature : any expression placed inside ${ } is **evaluated dynamically** and its result is substituted directly into the string.

Here are some examples

```typescript
const item = "Specimin"
const width = 2;

`Example of embedded expressions item=${item} Length=${4} Width=${width} Area=${4*width}.`
```

**PART 5: Numbers**

TypeScript has a single `number` type for all numeric values - integers, decimals, and negative numbers. 

Unlike some languages, there's no separate type for integers vs floating-point numbers.

```typescript
const integerVar = 42;
const decimalVar = 3.14;
const negativeVar = -10;

console.log(`   ${integerVar}       ${decimalVar}      ${negativeVar}  `);

console.log(` ${typeof integerVar}    ${typeof decimalVar}    ${typeof negativeVar} `);
```

**PART 6: Boolean Variables**

The `boolean` type has only two possible values: `true` or `false`. 

Booleans are commonly used for flags, conditions, and toggle states in your application.

Starting the name with `is` or `has` is a common convention.

```typescript
const isLoggedIn = false;
const hasPermission = true;

`Logged in: ${isLoggedIn}, Has permission: ${hasPermission}`
```

**PART 7: Boolean Values** 

In the Example below we use the builtin function Boolean to evaluate whether various values are true or false.


```typescript
const booleanValueOfOne = Boolean(1);

console.log(booleanValueOfOne);


console.log(`1 ${Boolean(1)}`);
console.log(`0 ${Boolean(0)}`);
console.log(`null ${Boolean(null)}`);
console.log(`undefined ${Boolean(undefined)}`);
console.log(`empty string ${Boolean('')}`);
console.log(`non-empty string ${Boolean('hello')}`);
```

Here are some basic JavaScript rules for coercing values into true or false.

- numbers   : 0 is false, other values are true    (NaN and -0 are false)
- null      : is false
- undefined : is false
- string    : empty string "" is false
            : non-empty string is true

(there are other rules, but these are the ones we are most likely to encounter)
