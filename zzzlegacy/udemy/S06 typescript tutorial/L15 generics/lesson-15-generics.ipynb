# Lesson 14: Generics

**PART 1: The problem** - code duplication:

```typescript
function getFirstString(array: string[]): string {
  return array[0];
}

function getFirstNumber(array: number[]): number {
  return array[0];
}

`${getFirstString(['a', 'b'])}, ${getFirstNumber([1, 2])} - Need separate functions!`
```

**Part 2: Generic function** - one function for any type:

```typescript
function getFirst<T>(array: T[]): T {
  return array[0];
}

const firstString = getFirst(['apple', 'banana', 'cherry']);
const firstNumber = getFirst([10, 20, 30]);

`${firstString}, ${firstNumber} - TypeScript infers types!`
```

**Part 3: Type parameters `<T>`:**

```typescript
function identity<T>(value: T): T {
  return value;
}

const stringResult = identity<string>('hello');
const numberResult = identity(42);

`${stringResult}, ${numberResult}`
```

**Part 4: Multiple type parameters:**

```typescript
function pair<T, U>(first: T, second: U): [T, U] {
  return [first, second];
}

const stringNumberPair = pair('Alice', 25);
const booleanStringPair = pair(true, 'active');

`[${stringNumberPair[0]}, ${stringNumberPair[1]}], [${booleanStringPair[0]}, ${booleanStringPair[1]}]`
```

**Part 5: Generic interfaces:**

```typescript
interface Box<T> {
  value: T;
  getValue: () => T;
}

const stringBox: Box<string> = {
  value: 'Hello',
  getValue: () => 'Hello',
};

const numberBox: Box<number> = {
  value: 42,
  getValue: () => 42,
};

`${stringBox.getValue()}, ${numberBox.getValue()}`
```

**Part 6: Arrays are generic:**

```typescript
const stringArray: Array<string> = ['a', 'b', 'c'];
const numberArray: Array<number> = [1, 2, 3];

const strings1: string[] = ['a', 'b'];
const strings2: Array<string> = ['a', 'b'];

'string[] is shorthand for Array<string>'
```
